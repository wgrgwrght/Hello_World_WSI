```python
# Extrct Patches
size_patch = 512

fixed_patch = patchextraction.get_patch_extractor(
    input_img=wsi,  
    method_name="slidingwindow",  
    patch_size=(
        size_patch,
        size_patch,
    ),  
    stride=(size_patch, size_patch),
)

mask_patch = patchextraction.get_patch_extractor(
    input_img=wsi_mask,  
    method_name="slidingwindow",  
    patch_size=(
        size_patch,
        size_patch,
    ),  
    stride=(size_patch, size_patch), 
)
```

# Sklearn Demo

```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import data
from skimage.color import rgb2hed, hed2rgb

i = 48301

# Example IHC image
ihc_rgb = fixed_patch[i]

# Separate the stains from the IHC image
ihc_hed = rgb2hed(ihc_rgb)

# Create an RGB image for each of the stains
null = np.zeros_like(ihc_hed[:, :, 0])
ihc_h = hed2rgb(np.stack((ihc_hed[:, :, 0], null, null), axis=-1))
ihc_e = hed2rgb(np.stack((null, ihc_hed[:, :, 1], null), axis=-1))
ihc_d = hed2rgb(np.stack((null, null, ihc_hed[:, :, 2]), axis=-1))

# Display
fig, axes = plt.subplots(2, 2, figsize=(7, 6), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].imshow(ihc_rgb)
ax[0].set_title("Original image")

ax[1].imshow(ihc_h)
ax[1].set_title("Hematoxylin")

ax[2].imshow(ihc_e)
ax[2].set_title("Eosin")  # Note that there is no Eosin stain in this image

ax[3].imshow(ihc_d)
ax[3].set_title("DAB")

for a in ax.ravel():
    a.axis('off')

fig.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import data
from skimage.color import rgb2hed, hed2rgb

from tiatoolbox.utils.misc import imread

i = 48301

# Example IHC image
ihc_rgb = imread('/content/drive/MyDrive/Patches/Patch_' + str(i) + '.png')

# Separate the stains from the IHC image
ihc_hed = rgb2hed(ihc_rgb)

# Create an RGB image for each of the stains
null = np.zeros_like(ihc_hed[:, :, 0])
ihc_h = hed2rgb(np.stack((ihc_hed[:, :, 0], null, null), axis=-1))
ihc_e = hed2rgb(np.stack((null, ihc_hed[:, :, 1], null), axis=-1))
ihc_d = hed2rgb(np.stack((null, null, ihc_hed[:, :, 2]), axis=-1))

# Display
fig, axes = plt.subplots(2, 2, figsize=(7, 6), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].imshow(ihc_rgb)
ax[0].set_title("Original image")

ax[1].imshow(ihc_h)
ax[1].set_title("Hematoxylin")

ax[2].imshow(ihc_e)
ax[2].set_title("Eosin")  # Note that there is no Eosin stain in this image

ax[3].imshow(ihc_d)
ax[3].set_title("DAB")

for a in ax.ravel():
    a.axis('off')

fig.tight_layout()
```

```python
from skimage.exposure import rescale_intensity

# Rescale hematoxylin and DAB channels and give them a fluorescence look
h = rescale_intensity(ihc_hed[:, :, 0], out_range=(0, 1),
                      in_range=(0, np.percentile(ihc_hed[:, :, 0], 99)))
d = rescale_intensity(ihc_hed[:, :, 2], out_range=(0, 1),
                      in_range=(0, np.percentile(ihc_hed[:, :, 2], 99)))

# Cast the two channels into an RGB image, as the blue and green channels
# respectively
zdh = np.dstack((null, d, h))

fig = plt.figure()
axis = plt.subplot(1, 1, 1, sharex=ax[0], sharey=ax[0])
axis.imshow(zdh)
axis.set_title('Stain-separated image (rescaled)')
axis.axis('off')
plt.show()
```

# Stain Separation

```python
from tiatoolbox.tools.stainextract import VahadaneExtractor, MacenkoExtractor, RuifrokExtractor

from tiatoolbox.utils.misc import imread

i = 48301

img = imread('/content/drive/MyDrive/Patches/Patch_' + str(i) + '.png')

extractor_v = VahadaneExtractor()
extractor_m = MacenkoExtractor()
extractor_r = RuifrokExtractor()

stain_matrix_v = extractor_v.get_stain_matrix(img)
stain_matrix_m = extractor_m.get_stain_matrix(img)
stain_matrix_r = extractor_r.get_stain_matrix(img)

plt.imshow(img)
```

```python
'''Using the stain augmentor stand alone'''

from tiatoolbox.tools.stainaugment import StainAugmentor

# Instantiate the stain augmentor and fit it on an image

stain_augmentor = StainAugmentor(stain_matrix=stain_matrix_r)

imh = stain_augmentor.apply(img)
```
