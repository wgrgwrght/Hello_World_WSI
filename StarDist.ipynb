# HoverNet

```python
# Define stain Normalizer
target_image = np.array(Image.open('/content/tiatoolbox/tiatoolbox/data/target_image.png'))
stain_normalizer = get_normalizer("vahadane")
stain_normalizer.fit(target_image)

def stain_norm_func(img):
    return stain_normalizer.transform(img)
```

```python
# Remove directory if already exists
!rm -rf /content/drive/MyDrive/Patches/sample_tile_results
```

```python
# Tried following models
#hovernet_original-consep (done)
#micronet_hovernet-consep (doesnt exist)
#hovernet_fast-pannuke (done)
#hovernet-original-kumar (doesnt exist)
#hovernet_fast-monusac (done)

# Define segmentation model
inst_segmentor = NucleusInstanceSegmentor(
    pretrained_model="hovernet_fast-pannuke",
    num_loader_workers=2,
    num_postproc_workers=2,
    batch_size=4,
)
ON_GPU = False

# Normalize
inst_segmentor.model.preproc_func = stain_norm_func

# Specify which image
num = 48301
img_file_name = '/content/drive/MyDrive/Patches/Patch_' + str(num) + '.png'
dir_save = "/content/drive/MyDrive/Patches/sample_tile_results/Patch_" + str(num)

# Make Prediction
tile_output = inst_segmentor.predict(
    [img_file_name],
    save_dir=dir_save,
    mode="tile",
    on_gpu=ON_GPU,
    crash_on_exception=True,
)

# Print number of found cells
tile_preds = joblib.load(f"{tile_output[0][1]}.dat")
print("Number of detected nuclei: {}".format(len(tile_preds)))
```

```python
# Reading the original image
tile_img = imread(img_file_name)

# defining the coloring dictionary: a dictionary that specifies a color to each class {type_id : (type_name, colour)}
color_dict = {
    0: ("neoplastic epithelial", (255, 0, 0)),
    1: ("Inflammatory", (255, 255, 0)),
    2: ("Connective", (0, 255, 0)),
    3: ("Dead", (0, 0, 0)),
    4: ("non-neoplastic epithelial", (0, 0, 255)),
}
 
# Create the overlay image
overlaid_predictions = overlay_prediction_contours(
    canvas=tile_img,
    inst_dict=tile_preds,
    draw_dot=False,
    type_colours=color_dict,
    line_thickness=2,
)

# showing processed results alongside the original images
fig = plt.figure()
ax1 = plt.subplot(1, 2, 1), plt.imshow(tile_img), plt.axis("off")
ax2 = plt.subplot(1, 2, 2), plt.imshow(overlaid_predictions), plt.axis("off")
```

# StarDist on H Channel

```python
# Separate the stains from the IHC image
ihc_hed = rgb2hed(fixed_patch[num])

# Create an RGB image for each of the stains
null = np.zeros_like(ihc_hed[:, :, 0])
ihc_h = hed2rgb(np.stack((ihc_hed[:, :, 0], null, null), axis=-1))

# Display
plt.subplot(1,2,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(ihc_h)
plt.axis("off")
plt.title("H-Channel")
plt.show()
```

```python
# Creates a pretrained model
model = StarDist2D.from_pretrained('2D_versatile_he')

# Predict
labels, _ = model.predict_instances(normalize(ihc_h), prob_thresh=0.3)

plt.subplot(1,3,1)
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.title("input image")

plt.subplot(1,3,2)
plt.imshow(render_label(labels, img=img))
plt.axis("off")
plt.title("prediction + input overlay")

plt.subplot(1,3,3)
plt.imshow(render_label(labels))
plt.axis("off")
plt.title("predictions")

print(str(len(_['points']))+" cells found!")
```
